```python
#probamos como sacar del df los datos que necesitaremos para añadir las marcas al mapa

for i in range(3):
       
    coord = df_albergues.iloc[i].WKT
    coord=coord.split()
    lon=float(coord[1].replace('(',''))
    lat=float(coord[2].replace(')',''))

    nombre = (df_albergues.iloc[i]['name']).title()
    web = (df_albergues.iloc[i].PaxinaWeb).split('\"')[1]
    web = web.split('?')[0]


    print(nombre)
    print(web)
    print(coord)
    print(lat)
    print(lon)
    print()

```

```python
#vamos a cargar también un shape con las Zonas de Especial Proteccion dos Valores Naturais

df_ZEPVN = gpd.read_file(ruta+'espacios protegidos/Rede_Galega_Espazos_Protexidos/ZEPVN.shp')
df_ZEPVN.sample(4)
```

```python


################################################################################# mapa
#creamos un objeto mapa de folium centrado en Galicia y con un tamaño y zoom inicial adecuados

lat = 42.77327229814228
lon = -7.7820556782640775
mapa = folium.Map(location=[lat, lon],zoom_start=8 ,width=1000, height=600, tiles=None)


################################################################################# bases
#añadimos capas de base o fondo

folium.TileLayer('CartoDB Positron', name='Mapa claro').add_to(mapa)
folium.TileLayer('Open Street Map', name='Mapa').add_to(mapa)
folium.TileLayer('Stamen Terrain', name='Relieve').add_to(mapa)
folium.TileLayer('https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}', name='Foto aérea', attr='Esri').add_to(mapa)



################################################################################# provivincias, concellos

#creamos un grupo al que añadiremos las entidades administrativas (así luego se añade al mapa las 3 juntas)
grupo_limites=folium.map.FeatureGroup(name='Concellos', overlay=True, control=True, show=True)

#creamos las funciones de estilo para cada capa vectorial del grupo
style_provincias = lambda x: {'color':'#3a6638', 'weight':4, 'fill':False}                          
style_concellos = lambda x: {'color':'#3a6638', 'weight':1, 'fillOpacity':0.1}

#añadimos al grupo las capas
folium.GeoJson(df_concellos, name='Concellos', tooltip=folium.GeoJsonTooltip(['Concello']), style_function = style_concellos).add_to(grupo_limites)
    #a esta capa le hemos puesto un tooltip que muestra el nombre del concello correspondiente en función de donde esté el puntero
folium.GeoJson(df_provincias, name='Provincias', style_function = style_provincias).add_to(grupo_limites)

#añadimos el grupo al mapa
grupo_limites.add_to(mapa)


################################################################################# caminos

#creamos función de estilo para los caminos
style_camino = lambda x: {'color':'#5e3c00', 'weight':4}

#añadimos los caminos al mapa, con el nombre de cada uno como tooltip
folium.GeoJson(df_camino_frances, name='Camino francés', tooltip='Camino francés',style_function = style_camino).add_to(mapa)
folium.GeoJson(df_camino_ingles, name='Camino inglés', tooltip='Camino inglés', style_function = style_camino).add_to(mapa)
folium.GeoJson(df_camino_norte_costa, name='Camino norte costa', tooltip='Camino norte costa', style_function = style_camino).add_to(mapa)
folium.GeoJson(df_camino_norte_interior, name='Camino norte interior', tooltip='Camino norte interior', style_function = style_camino).add_to(mapa)
folium.GeoJson(df_camino_portugues, name='Camino portugués', tooltip='Camino portugués', style_function = style_camino).add_to(mapa)
folium.GeoJson(df_camino_via_plata, name='Camino vía plata', tooltip='Camino vía plata', style_function = style_camino).add_to(mapa)


############################################################################## albergues

#creamos un cluster de marcas
marker_cluster_albergues = folium.plugins.MarkerCluster(name='Albergues del camino de Santiago')

#iteramos el df de los albergues sacando las coordenadas y los datos para el popup
for i in range(len(df_albergues)):
    try:            
        coord = df_albergues.iloc[i].WKT
        coord=coord.split()
        lon=float(coord[1].replace('(',''))
        lat=float(coord[2].replace(')',''))

        nombre = (df_albergues.iloc[i]['name']).title()
        web = (df_albergues.iloc[i].PaxinaWeb).split('\"')[1]
        web = web.split('?')[0]

        
        #creamos un popup con script html para poder meter el link activo que funcione
        pp= folium.Html('<p>'+nombre+'</p><a href="'+ web +'"target="_blank">Link a la web</a>', script=True)
        popup = folium.Popup(pp)

        #añadimos al cluster la marca de cada albergue
        folium.Marker([lat, lon], popup=popup, icon=folium.Icon(prefix='fa', icon='home', color='green')).add_to(marker_cluster_albergues)      
    except:
        pass

#añadimos el cluster al mapa
marker_cluster_albergues.add_to(mapa)


#añadimos también una marca para la Cateral de Santiago
folium.Marker([42.88057976730063, -8.544896294471274], popup='Catedral de Santiago', icon=folium.Icon(prefix='fa', icon='plus', color='darkred')).add_to(mapa)


############################################################################## ZEPVN

style_zepvn = lambda x: {'color':'#7fa838', 'weight':1, 'fillOpacity':0.5}

#añadimos las Zonas de especial protección dos valores naturais al mapa, por defecto apagada
#como tooltip ponemos el nombre correspondiente y la superficie (Ha)
folium.GeoJson(df_ZEPVN, name='Zonas de Especial Protección dos Valores Naturais', show=False, style_function = style_zepvn, tooltip=folium.GeoJsonTooltip(['NOME','SUP_Has'],['ZEPVN:','Hectáreas:'])).add_to(mapa)



############################################################################## controlador de capas

#añadimos el controlador de capas
folium.LayerControl().add_to(mapa)

```
